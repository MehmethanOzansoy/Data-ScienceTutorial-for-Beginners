```python

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns  # visualization tool
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
data = pd.read_csv('/kaggle/input/pokemon-dataset-with-team-combat/pokemon.csv')
data.head()
```

```python
#satır ve sütüunların özellikleri
data.info()
```

```python
#verilerin korelasyonları veriliyor
#Korelasyon, olasılık kuramı ve istatistikte iki rassal değişken arasındaki doğrusal 
#ilişkinin yönünü ve gücünü belirtir. Genel istatistiksel kullanımda korelasyon, 
#bağımsızlık durumundan ne kadar uzaklaşıldığını gösterir. Farklı durumlar için 
#farklı korelasyon katsayıları geliştirilmiştir.
data.corr()
```

```python
f,ax = plt.subplots(figsize=(9, 9))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
#korelasyon haritası
```

```python
data.Speed.plot(kind = 'line', color = 'b',label = 'hız',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Defense.plot(color = 'r',label = 'savunması',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Line Plot')            # title = title of plot
plt.show()
```

```python
data.Speed.plot(kind = 'line', color = 'y',label = 'speed',linewidth=2.5 ,alpha = 0.5,grid = True,linestyle = 'solid')# '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red')
plt.xlabel('Attack')              # label = name of label
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')   
plt.show()
```

```python
data.Speed.plot(kind = 'hist',color = 'r',bins = 50,figsize = (5,5))
#kind : str
#‘line’ : line plot (default)
#‘bar’ : vertical bar plot
#‘barh’ : horizontal bar plot
#‘hist’ : histogram
#‘box’ : boxplot
#‘kde’ : Kernel Density Estimation plot
#‘density’ : same as ‘kde’
#‘area’ : area plot
#‘pie’ : pie plot
#‘scatter’ : scatter plot
#‘hexbin’ : hexbin plot
plt.show()
```
